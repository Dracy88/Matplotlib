# Scatterplots

```python
import numpy as np

x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.figure()
# plot a data series 'Teenager people' in red using the first two elements of x and y
plt.scatter(x[:3], y[:3], s=100, c='red', label='People with less of 14 years')
# plot a second data series 'Mature people' in blue using the last three elements of x and y 
plt.scatter(x[3:], y[3:], s=100, c='blue', label='People with atleast 14 years')

# add a label to the x axis
plt.xlabel('The number of times that the subject has kicked a ball')
# add a label to the y axis
plt.ylabel('The age of the people')
# add a title
plt.title('Relationship between ball kicking and ages')

# add the legend to loc=0 (the top left hand corner), also gets rid of the frame and adds a title
plt.legend(loc=0, frameon=False, title='Legend')
```
